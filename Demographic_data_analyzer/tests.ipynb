```python
import pandas as pd

df = pd.read_csv('adult.data.csv')
df.head()
```

```python
race_count = pd.Series(df['race'].value_counts(), dtype='int16')
print('Number of each race:\n', race_count)
```

```python
avg_age_men = round(df[df['sex'] == 'Male']['age'].mean(),1)
avg_age_men
```

```python
# What is the percentage of people who have a Bachelor's degree?
# perc_bachelors_deg = df[df['education'] == ['Bachelors', 'Masters','Doctorate']]

total_education = df['education'].value_counts().sum()
bachelors = df[df['education'] == 'Bachelors']['education'].value_counts().sum()

print(round(bachelors/total_education * 100,1))
```

```python
# What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
# What percentage of people without advanced education make more than 50K?
    
# with and without `Bachelors`, `Masters`, or `Doctorate`
higher_above_50K = df[(df['education'].isin(['Bachelors','Masters','Doctorate'])) & (df['salary'] == '>50K')].value_counts().sum()
higher_education = df[df['education'].isin(['Bachelors','Masters','Doctorate'])].value_counts().sum()
lower_above_50K = df[~(df['education'].isin(['Bachelors','Masters','Doctorate'])) & (df['salary'] == '>50K')].value_counts().sum()
lower_education = df[~df['education'].isin(['Bachelors','Masters','Doctorate'])].value_counts().sum()

# percentage with salary >50K
higher_education_rich = round(higher_above_50K /higher_education*100,1)
lower_education_rich = round(lower_above_50K/lower_education*100,1)
print([higher_education_rich, lower_education_rich])
```

```python
# What is the minimum number of hours a person works per week (hours-per-week feature)?
min_work_hours = df['hours-per-week'].min()
```

```python
# What percentage of the people who work the minimum number of hours per week have a salary of >50K?
num_min_workers = df[(df['hours-per-week'] == min_work_hours) & (df['salary'] == '>50K')].value_counts().sum()
num_min_workers
```

```python
rich_percentage = num_min_workers / df[df['hours-per-week'] == min_work_hours].value_counts().sum() * 100
rich_percentage
```
